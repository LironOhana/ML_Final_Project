```python
from nba_api.stats.endpoints import playercareerstats
import pandas as pd
from nba_api.stats.static import players
from nba_api.stats.static import players
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Players stats

```python
players = players.get_players()

# Convert active players data into a DataFrame
players_df = pd.DataFrame(players)

# Display the DataFrame
players_df=players_df[['id','full_name']]
players_df
```

```python
career_stats_list = []
for index, row in players_df.iterrows():
    player_id = row['id']
    career = playercareerstats.PlayerCareerStats(player_id=str(player_id)) 
    career_stats_dict = career.get_dict()
    player_stats = career_stats_dict['resultSets'][0]['rowSet']
    

    career_stats_list.extend(player_stats)
```

```python
columns = career_stats_dict['resultSets'][0]['headers']
result_df = pd.DataFrame(career_stats_list, columns=columns)
result_df
```

```python
# Assuming 'id' is the common column between players_df and result_df
result_df['full_name'] = result_df['PLAYER_ID'].map(players_df.set_index('id')['full_name'])

result_df
```

### All Star players Crawling

```python
# Initialize an empty list to store the data
data = []
year_text = None  # Initialize year_text outside the loop

# Fetch the HTML content
url = 'https://www.justallstar.com/nba-all-star-game/player-lists/rosters/#1547322049361-993feb14-2aa3'
response = requests.get(url)
html_content = response.text

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')
# Extract data from td tags and append to the list
for index, td in enumerate(soup.find_all('td')):
    td_text = td.get_text().strip()
    
    if index % 2 == 0:  # Every second td
        if td.a:  # Check if td contains an 'a' tag
            year_text = td.a.get_text()  # Update year_text if it's a year
        else:
            data.append({'Player': td_text, 'Year': year_text})

# Create a DataFrame from the list of dictionaries
all_stars_df = pd.DataFrame(data)
```

```python
all_stars_df = all_stars_df[(all_stars_df['Player'] != '')&(all_stars_df['Player'] != 'EAST')&(all_stars_df['Player'] != 'WEST')&(~all_stars_df['Player'].str.contains('Team'))]
all_stars_df['Year'].fillna('2019', inplace=True)

# Reset the index after filtering rows
all_stars_df.reset_index(drop=True, inplace=True)

all_stars_df
```

### Add id column

```python
# Merge all_stars_df with players_df to get the corresponding IDs
all_stars_with_ids = pd.merge(all_stars_df, players_df[['id', 'full_name']], how='left', left_on='Player', right_on='full_name')

# Drop the redundant 'full_name' column
all_stars_with_ids.drop('full_name', axis=1, inplace=True)
all_stars_with_ids['id'] = all_stars_with_ids['id'].astype(str).replace('\.0', '', regex=True)
```

```python
# Remove rows with 'NUN' in the 'id' column
all_stars_with_ids = all_stars_with_ids[all_stars_with_ids['id'] != 'nan']

# Remove duplicates
all_stars_with_ids = all_stars_with_ids.drop_duplicates()

# Display the resulting DataFrame
all_stars_with_ids
```

### Crawling All Star stats

```python
from nba_api.stats.endpoints import playercareerstats
career_stats_list1 = []

for index, row in all_stars_with_ids.iterrows():

    player_id = row['id']
    career = playercareerstats.PlayerCareerStats(player_id=str(player_id)) 
    career_stats_dict = career.get_dict()
    player_stats = career_stats_dict['resultSets'][0]['rowSet']
    

    career_stats_list1.extend(player_stats)
```

```python
columns = career_stats_dict['resultSets'][0]['headers']
allstars_stats = pd.DataFrame(career_stats_list1, columns=columns)
allstars_stats
```

### Combine the data frames

```python
# Assuming both DataFrames have the same columns
combined_df = pd.concat([result_df, allstars_stats], ignore_index=True)

combined_df = combined_df.drop_duplicates(subset=['PLAYER_ID', 'SEASON_ID','TEAM_ID'])

# Print the DataFrame after removing duplicates
combined_df
```

```python
# Assuming 'id' is the common column between players_df and result_df
combined_df['full_name'] = combined_df['PLAYER_ID'].map(players_df.set_index('id')['full_name'])
```

### Fix Year column

```python
combined_df['Year'] = combined_df['SEASON_ID'].str.split('-').str[0]

combined_df
```

### Add All Star column

```python
# Create a new column named 'All_Star' in result_df
combined_df['All_Star'] = 0

# For each player in all_stars_df
for index, row in all_stars_df.iterrows():
    player_name = row['Player']
    year = row['Year']
    
    # Check if there is a player with the same name and year in result_df
    mask = (combined_df['full_name'] == player_name) & (combined_df['Year'] == year)
    combined_df.loc[mask, 'All_Star'] = 1

combined_df[combined_df['All_Star'] == 1]
combined_df.head()
```
